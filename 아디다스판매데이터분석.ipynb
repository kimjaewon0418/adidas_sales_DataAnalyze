# 데이터 로드 및 확인

```python
import pandas as pd
```

```python
df = pd.read_excel('/content/Adidas US Sales Datasets (1).xlsx')
```

```python
df
```

# 데이터 전처리

```python
af = df.iloc[4:,1:]
```

```python
af.rename(columns={'Unnamed: 1':'Retailer'}, inplace=True)
af.rename(columns={'Unnamed: 2':'Retailer ID'}, inplace=True)
af.rename(columns={'Unnamed: 3':'Invoice Date'}, inplace=True)
af.rename(columns={'Unnamed: 4':'Region'}, inplace=True)
af.rename(columns={'Unnamed: 5':'State'}, inplace=True)
af.rename(columns={'Unnamed: 6':'City'}, inplace=True)
af.rename(columns={'Unnamed: 7':'Product'}, inplace=True)
af.rename(columns={'Unnamed: 8':'Price per Unit'}, inplace=True)
af.rename(columns={'Unnamed: 9':'Units Sold'}, inplace=True)
af.rename(columns={'Unnamed: 10':'Total Sales'}, inplace=True)
af.rename(columns={'Unnamed: 11':'Operating Profit'}, inplace=True)
af.rename(columns={'Unnamed: 12':'Operating Margin'}, inplace=True)
af.rename(columns={'Unnamed: 13':'Sales Method'}, inplace=True)
```

```python
af
```

```python
af['Sales Method'].unique()
```

```python
af.info()
```

```python
af['Price per Unit'] = af['Price per Unit'].astype('float')
# 만약 숫자앞에 가격을 나타내는 특수문자($,%, ,)가 포함되어 있을 경우에 특수문자 제거하는 법
# af['Price per Unit'] = af['Price per Unit'].str.replace('[$%,]','').astype('float')
```

```python
af['Units Sold'] = af['Units Sold'].astype('float')
af['Total Sales'] = af['Total Sales'].astype('float')
af['Operating Profit'] = af['Operating Profit'].astype('float')
af['Operating Margin'] = af['Operating Margin'].astype('float')
```

```python
af.head(1)
```

```python
af['Invoice Date'] = pd.to_datetime(af['Invoice Date'], format = '%Y-%m-%d')
```

```python
af['year'] = af['Invoice Date'].dt.year
af['month'] = af['Invoice Date'].dt.month
```

```python
af.head()
```

```python
af['Retailer ID'] = af['Retailer ID'].astype('string')
```

```python
af.isnull().sum() #결측치 없음
```

```python
af.info()
```

# 월별로 상점의 숫자는 어떻게 될까?
2020년에는 Foot Locker의 상점 수가 1개였는데 2021년에는 3개로 늘지 않았을까?

```python
af[['Retailer', 'Retailer ID','year','month']].drop_duplicates().sort_values('Retailer')
```

```python
af.pivot_table(index = ['year','month'], values = 'Retailer ID', aggfunc = 'count')
```

```python
af.pivot_table(index = ['year','month'], columns='Retailer', values = 'Retailer ID', aggfunc = 'count')
```

```python
af.pivot_table(index = ['year','month'], columns='Retailer', values = 'Retailer ID', aggfunc = 'count').plot(kind = 'area', figsize=(20,7))
```

확인해보니 코로나로인해 매출이 감소 되었다가 2021년 초부터 백신이 개발되어 유통이 늘어나 판매량이 증가하였다.
